```
# Questions categorized by topics
questions_by_topic = {
    "General Knowledge": [
        ("What does AI stand for?", "Artificial Intelligence"),
        ("What is the capital of France?", "Paris"),
        ("What is 5 + 3?", "8"),
        ("Who wrote 'Hamlet'?", "Shakespeare")
    ],
    "Python Programming": [
        ("What is the output of 2 + 2?", "4"),
        ("Which library is used for data visualization?", "matplotlib"),
        ("What does ML stand for?", "Machine Learning"),
        ("What keyword is used to define a function in Python?", "def")
    ],
    "Data Science": [
        ("What does CSV stand for?", "Comma Separated Values"),
        ("Which library is used for data manipulation?", "pandas"),
        ("What is the full form of EDA?", "Exploratory Data Analysis"),
        ("What does NaN stand for?", "Not a Number")
    ]
}

def chatbot():
    print("ü§ñ: Hello! I'm your chatbot. My name is ChatBuddy.")
    print("ü§ñ: You can choose a topic and answer questions. After four questions, you can decide to continue or stop.")
    print("ü§ñ: Type 'exit' anytime to quit.\n")

    # Main chatbot loop
    while True:
        # Greet user
        name = input("ü§ñ: What's your name? ").strip()
        if name.lower() == "exit":
            print("ü§ñ: Goodbye! üëã")
            break

        print(f"ü§ñ: Nice to meet you, {name}!")

        while True:
            # Display available topics
            print("\nü§ñ: Choose a topic:")
            for i, topic in enumerate(questions_by_topic.keys(), 1):
                print(f"  {i}. {topic}")

            # User selects a topic
            choice = input("Your choice (1/2/3): ").strip()
            if choice.lower() == "exit":
                print("ü§ñ: Goodbye! üëã")
                return

            try:
                topic = list(questions_by_topic.keys())[int(choice) - 1]
            except (ValueError, IndexError):
                print("ü§ñ: Invalid choice. Please select a valid option.")
                continue

            print(f"\nü§ñ: You selected '{topic}'. Let's start!\n")

            # Ask 4 questions from the selected topic
            questions = questions_by_topic[topic]
            correct_answers = 0

            for i, (question, correct_answer) in enumerate(questions[:4], 1):
                print(f"Question {i}: {question}")
                user_answer = input("Your Answer: ").strip()

                if user_answer.lower() == "exit":
                    print("ü§ñ: Goodbye! üëã")
                    return

                if user_answer.lower() == correct_answer.lower():
                    print("‚úÖ Correct!")
                    correct_answers += 1
                else:
                    print(f"‚ùå Incorrect. The correct answer is: {correct_answer}")

            print(f"\nü§ñ: You answered {correct_answers}/4 questions correctly.")

            # Ask if the user wants to continue
            continue_choice = input("ü§ñ: Do you want to continue? (yes/no): ").strip().lower()
            if continue_choice == "no":
                print("ü§ñ: Thank you for playing! Have a great day! üëã")
                return
            elif continue_choice != "yes":
                print("ü§ñ: Invalid response. Exiting.")
                return

# Run the chatbot
chatbot()
```
